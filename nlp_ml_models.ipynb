```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
from sklearn.model_selection import train_test_split
tr, valid = train_test_split(train, random_state=42, test_size=0.30, shuffle=True , stratify = train['Quantitative Finance'])
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(max_features= 5000)
x_train_1 = vectorizer.fit_transform(tr['TITLE'])
x_valid_1 = vectorizer.transform(valid['TITLE'])
x_test_1 = vectorizer.transform(test['TITLE'])
x_train_2 = vectorizer.fit_transform(tr['ABSTRACT'])
x_valid_2 = vectorizer.transform(valid['ABSTRACT'])
x_test_2 = vectorizer.transform(test['ABSTRACT'])
```

```python
X_tr = pd.concat([pd.DataFrame(x_train_1.toarray()),pd.DataFrame(x_train_2.toarray() , columns = range(5000,10000))], axis = 1)
X_valid = pd.concat([pd.DataFrame(x_valid_1.toarray()),pd.DataFrame(x_valid_2.toarray() , columns = range(5000,10000))], axis = 1)
X_test = pd.concat([pd.DataFrame(x_test_1.toarray()),pd.DataFrame(x_test_2.toarray() , columns = range(5000,10000))], axis = 1)
```

```python
y_col = train.columns[2:8]
```

```python
y_train = tr.select_dtypes([int])[y_col]
y_valid = valid.select_dtypes([int])[y_col]
y_test = y_test.select_dtypes([int])[y_col]
```

```python
import skmultilearn
from sklearn import svm
from skmultilearn.problem_transform import BinaryRelevance
from skmultilearn.problem_transform import ClassifierChain
from sklearn.naive_bayes import MultinomialNB
from sklearn.naive_bayes import MultinomialNB
from skmultilearn.problem_transform import LabelPowerset
from sklearn.linear_model import LogisticRegression
from sklearn.multiclass import OneVsRestClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
from sklearn.linear_model import RidgeClassifier
from sklearn.linear_model import BayesianRidge
from sklearn.neighbors import KNeighborsClassifier
```

```python

# initialize binary relevance multi-label classifier
# with a gaussian naive bayes base classifier
all_algo = pd.DataFrame(columns = ['train_ac','test_ac','train_f1','test_f1'])

algos = [LogisticRegression,MultinomialNB,RidgeClassifier]
for x in algos:
    if x == MultinomialNB:
        classifier = LabelPowerset(x())
        classifier.fit(X_tr, y_train)
        predictions = classifier.predict(X_valid)
        tr_pre = classifier.predict(X_tr)
        test_ac = accuracy_score(y_valid,predictions)
        train_ac = accuracy_score(y_train,tr_pre)
        test_f1 = f1_score(predictions, y_valid , average='micro')
        train_f1 = f1_score(tr_pre,y_train, average='micro')
        print([x,train_ac,test_ac,train_f1,test_f1])
    else:
        classifier = LabelPowerset(x(max_iter= 5000))
        classifier.fit(X_tr, y_train)
        predictions = classifier.predict(X_valid)
        tr_pre = classifier.predict(X_tr)
        test_ac = accuracy_score(y_valid,predictions)
        train_ac = accuracy_score(y_train,tr_pre)
        test_f1 = f1_score(predictions, y_valid , average='micro')
        train_f1 = f1_score(tr_pre,y_train, average='micro')
        print([x,train_ac,test_ac,train_f1,test_f1])
```

```python
classifier=LabelPowerset(MultinomialNB(alpha=0.1))
previous_score = 0
for alpha in np.arange(0,1,0.1):
    
    sub_classifier = LabelPowerset(MultinomialNB(alpha=alpha))
    sub_classifier.fit(X_tr, y_train)
    predictions = sub_classifier.predict(X_valid)
    score = f1_score(y_valid,predictions , average='micro')
    if score>previous_score:
        classifier=sub_classifier
    previous_score = score
    print("Alpha: {}, Score : {}".format(alpha,score))
```
